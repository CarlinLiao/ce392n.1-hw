```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[12,5,12,9,8]
possible = np.arange(0,12.1,0.1)
plt.plot(
    possible,
    [np.prod([np.exp(-l)*(l**xi)/np.math.factorial(xi) for xi in x]) for l in possible]
)
plt.minorticks_on()
plt.grid(b=True, which='major')
plt.grid(b=True, which='minor', color='lightgray', linestyle='--')
plt.ylabel('L(λ)')
plt.xlabel('λ')
plt.tight_layout()
plt.savefig('hw2.png', dpi='figure')
plt.show()
```
